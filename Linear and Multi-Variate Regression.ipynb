## Simple Linear Regression

A simple linear regression model estimates the relationship between two quantitative variables where one is referred to as the independent variable and the other the dependent variable. The independent variable (X) is used to predict and also called the predictor while the predicted variable is referred to as the response variable (Y) (e.g. finding the relationship between the amount of CO2 gas emitted and the number of trees cut down). The value of Y can be obtained from X by finding the line of best fit (regression line) with minimum error for the data points on a scatter plot for both variables. The UCI Machine Learning Repository: Energy efficiency Data Set is used in this module for better understanding of the concepts. We select a sample of the dataset and use the relative compactness column as the predictor and the heating load column the response variable.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

We shall read our Energy Efficiency data from the UCI Machine Learning Repository using pandas for excel file reading

```python
# reading data and looking at the first 10 entries

data = pd.read_excel('https://archive.ics.uci.edu/ml/machine-learning-databases/00242/ENB2012_data.xlsx')
data.head(10)
```

```python
# we shall rename the columns to better understand our data 

data.columns = data.columns.map({'X1':'Relative_Compactness', 'X2': 'Surface_Area', 
                'X3':  'Wall_Area', 'X4': 'Roof_Area', 'X5': 'Overall_Height',
                'X6': 'Orientation', 'X7': 'Glazing_Area', 
                'X8': 'Glazing_Area_Distribution', 
                'Y1': 'Heating_Load', 'Y2': 'Cooling_Load'})

data.head()
```

```python
# we now select a sample of the dataset

simple_linear_reg_data = data[['Relative_Compactness', 'Cooling_Load']].sample(20, random_state=3)
```

```python
# Plotting a simple linear regression of the above sample using seaborn 

sns.regplot(x = "Relative_Compactness", y = 'Cooling_Load', data = simple_linear_reg_data)
```

### Another Simple Linear Regression

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets, linear_model
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
# Create a data set for analysis
x, y = make_regression(n_samples=500, n_features = 1, noise=25, random_state=0)
```

```python
# Split the data set into testing and training data
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
# Create a linear regression object
regression = linear_model.LinearRegression()
```

```python
# Train the model using the training set
regression.fit(x_train, y_train)
```

```python
# Make predictions using the testing set
y_predictions = regression.predict(x_test)
```

```python
# Plot the data
sns.set_style("darkgrid")
sns.regplot(x_test, y_test, fit_reg=False)
plt.plot(x_test, y_predictions, color='black')
```

```python
# Remove ticks from the plot
plt.xticks([])
plt.yticks([])

plt.tight_layout()
plt.show()
```

## Multiple Linear Regression (Multi-Variate)

We will run this using our initial Energy Efficiency Dataset. 
In this section, we will develop a multivariate multiple regression model to study the effect of eight input variables on two output variables, which are the heating load and the cooling load, of residential buildings.

```python
# Plotting the correlation matrix heatmap

sns.heatmap(data.corr(), annot = True, vmin = -1, vmax = 1, fmt = '.2g')
plt.show()
```

It is seen that some variables highly correlated with others, such as surface are, roof area, highly correlating with relative compactness 


```python
# we begin our modelling by normalizing our dataset to a common scale using the min max scaler

from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()
norm_data = pd.DataFrame(scaler.fit_transform(data), columns = data.columns)
#then we look at our normalized data
norm_data.head()
```

```python
# Now we specify our features and target variable in order to proceed with the modelling. 
# Our target variable for now would be the Heating_Load. This means we will drop the column from the rest of our dataset to get our input features x 
# Lets call the target variable 'heating_target'

features_data = norm_data.drop(columns = ['Heating_Load', 'Cooling_Load'])
heating_target = norm_data['Heating_Load']
```

```python
# Now we split our dataset into the training and testing sets. This will be in the ratio 70:30

from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(features_data, heating_target, test_size = 0.3, random_state = 1)
```

```python
print(x_train.info(), '/n', x_test.info())
```

```python
print(y_train, '/n', y_test)
```

```python
# Now defining our linear model 

from sklearn.linear_model import LinearRegression 

linear_model = LinearRegression()

# fitting the model to the training data 
linear_model.fit(x_train, y_train)

# obtaining the predictions 
predict_values = linear_model.predict(x_test)
```

```python
# Calculating the Mean Absolute Error 

from sklearn.metrics import mean_absolute_error
mae = round(mean_absolute_error(y_test, predict_values), 3)
mae
```

```python
# Calculating the R-Squared value 

from sklearn.metrics import r2_score 
rsquared = round(r2_score(y_test, predict_values), 3)
rsquared
```

```python
# Calculating the residual sum of squares 

import numpy as np 
rss = round(np.sum(np.square(y_test - predict_values)), 3)
rss
```

```python
# Calculating the root mean square error 

from sklearn.metrics import mean_squared_error 

rmse = round(np.sqrt(mean_squared_error(y_test, predict_values)), 3)
rmse
```

### Comparing the Effects of Regularization 

```python
def get_weights_data(linear_model, feat, col_name):
    # this function returns the weight of every feature 
    weights = pd.Series(linear_model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df 

# Feature selection and Lasso Regression 
from sklearn.linear_model import Lasso 
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(x_train, y_train)

# Feature selection and Ridge regression 
from sklearn.linear_model import Ridge 
ridge_reg = Ridge(alpha = 0.5)
ridge_reg.fit(x_train, y_train) 

linear_model_weights = get_weights_data(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_data(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_data(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
final_weights
```
