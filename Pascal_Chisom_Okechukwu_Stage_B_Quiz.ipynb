```python
# We will start off this notebook by importing the basic, necessary libraries 

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
```

```python
# Setting up our data and looking at the first few rows

data = pd.read_csv('energydata_complete.csv')
data.head()
```

### Running a Simple Linear Regression Model to define the relationship between the 'temperature in the living room in Celsius' (x = T2) and the 'temperature outside the building (y = T6)' 

```python
# The Simple Linear Regression 

x_simple = pd.DataFrame(data['T2'])
y_simple = data['T6']

x_train_simple, x_test_simple, y_train_simple, y_test_simple = train_test_split(x_simple, y_simple, test_size = 0.3, random_state = 42)

simp_linear_mod = LinearRegression()
simp_linear_mod.fit(x_train_simple, y_train_simple)

simp_linear_pred = simp_linear_mod.predict(x_test_simple)
```

```python
# Now calculating the R-Squared value for the model 

from sklearn.metrics import r2_score 

print('The R-Squared value for this modoel is ', round(r2_score(y_test_simple, simp_linear_pred), 2))
```

### Running the Multi-Variate Linear Regression where the target variable is 'Appliances' 

```python
# To start off, we will drop the 'date' and 'lights' columns before normalizing the dataset 

data = data.drop(columns = ['date', 'lights'])
data.head()
```

```python
# Inspecting our data for cleanliness 

print(data.info())
```

```python
# Our dataset is very clean, with no missing values. Therefore we move on to normalizing it, using the MinMaxScaler from sklearn 

from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()
norm_data = pd.DataFrame(scaler.fit_transform(data), columns = data.columns)
```

```python
# Inspecting our dataset to confirm that normalization worked 

norm_data.head()
```

```python
# Now we define our features data and the target variable. Remember the target variable from the problem is 'Appliances'
# In doing this, we will drop the 'Appliances' column from the rest of the dataset (the latter which will then be our features data)
# while the former becomes our target variable

features_data = norm_data.drop(columns = ['Appliances'])
appliance_target = norm_data['Appliances']
```

```python
# At this stage, we will split our data into the training and testing factions 
# The train_test_split library has been imported already 

x_train, x_test, y_train, y_test = train_test_split(features_data, appliance_target, test_size = 0.3, random_state = 42)
```

```python
# Inspecting our Features data training and test set

print(x_train.info(), '/n', x_test.info())
```

```python
# Inspecting our Target Variable training and testing set 

print(y_train, '/n', y_test)
```

```python
# Now running the Multi-Linear Regression

linear_model = LinearRegression()

# fitting the model to the training data 
linear_model.fit(x_train, y_train)

# obtaining the predictions 
predict_values = linear_model.predict(x_test)
```

### Model Evaluation

```python
# Calculating the Mean Absolute Error 

from sklearn.metrics import mean_absolute_error
mae = round(mean_absolute_error(y_test, predict_values), 2)
mae
```

```python
# Calculating the residual sum of squares 

rss = round(np.sum(np.square(y_test - predict_values)), 2)
rss
```

```python
# Calculating the root mean square error 

from sklearn.metrics import mean_squared_error 

rmse = round(np.sqrt(mean_squared_error(y_test, predict_values)), 3)
rmse
```

```python
# Calculating the R-Squared value, otherwise known as the Coefficient of Determination

from sklearn.metrics import r2_score 
rsquared = round(r2_score(y_test, predict_values), 2)
rsquared
```

```python
# Estimating the Parameters for our Linear Regression Model 

feature_weights = pd.DataFrame(linear_model.coef_, index = features_data.columns, columns = ['Coefficient Value'])
feature_weights.sort_values('Coefficient Value')
```

```python
# Feature selection and Ridge regression 

from sklearn.linear_model import Ridge 
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(x_train, y_train) 

ridge_predict = ridge_reg.predict(x_test)
```

```python
# new Model evaluation 

print('New Test rmse is ', round(np.sqrt(mean_squared_error(y_test, ridge_predict)), 3))
```

```python
# Feature selection and Lasso Regression 

from sklearn.linear_model import Lasso 
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(x_train, y_train)
lasso_predict = lasso_reg.predict(x_test)
new_train_lasso = lasso_reg.predict(x_train)
```

```python
# Parameters evaluation for the Lasso model 

coeff = pd.DataFrame(lasso_reg.coef_, index = features_data.columns, columns = ['Value'])
coeff.sort_values('Value')
```

```python
# new Model evaluation 

print('New Test rmse via Lasso reg. is ', round(np.sqrt(mean_squared_error(y_test, lasso_predict)), 3))
```
